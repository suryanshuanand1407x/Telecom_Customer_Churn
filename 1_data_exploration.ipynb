```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/Users/suryanshu/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Display basic information about the dataset
print("Dataset Information:")
print(df.info())  # Overview of dataset structure, nulls, and datatypes

print("\nFirst 5 Rows:")
print(df.head())  # Preview the first few rows of data

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())

# Basic statistics of numerical features
print("\nStatistical Summary of Numerical Columns:")
print(df.describe())

# Explore the target column (Churn distribution)
print("\nChurn Distribution:")
print(df['Churn'].value_counts())
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()

# Check unique values in each column (to identify categorical features)
print("\nUnique Values in Each Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```
