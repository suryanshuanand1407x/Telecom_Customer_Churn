```python
# Import necessary libraries
import pandas as pd

# Load the dataset
df = pd.read_csv('/Users/suryanshu/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# 1. Handle Missing Values
print("Missing Values Before Cleaning:")
print(df.isnull().sum())

# Drop rows or columns with many missing values
df = df.dropna()  # Drop rows with missing values
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# 2. Drop Irrelevant Columns
# For example, 'customerID' is not useful for prediction
df = df.drop(['customerID'], axis=1)

# 3. Encode Categorical Variables
# Convert binary categorical variables to 0 and 1
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# One-hot encode categorical variables with more than two categories
df = pd.get_dummies(df, drop_first=True)

# 4. Normalize or Standardize Numerical Columns (if needed)
# Scaling can be added here if numeric features vary greatly in scale

# Save the cleaned dataset for further use
df.to_csv('../data/processed/churn_processed.csv', index=False)

print("\nPreprocessed Data (First 5 Rows):")
print(df.head())

print("\nShape of Dataset After Preprocessing:")
print(df.shape)
```

```python
# Import necessary libraries
import pandas as pd
import joblib

# Load the preprocessed dataset
df = pd.read_csv('/Users/suryanshu/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Ensure all features are numeric
X = df.drop('Churn', axis=1)  # Drop the target column

# Encode categorical variables
X = pd.get_dummies(X, drop_first=True)

# Keep only numeric columns
X = X.select_dtypes(include=[float, int])

# Save the trained features
trained_features = X.columns.tolist()
joblib.dump(trained_features, '../data/processed/trained_features.pkl')
```
