```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed dataset
df = pd.read_csv('/Users/suryanshu/Downloads/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# 1. Check Churn Distribution
print("Churn Distribution:")
print(df['Churn'].value_counts())
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()

# 2. Correlation Matrix
# Select only numeric columns to calculate correlations
numeric_df = df.select_dtypes(include=[float, int])

plt.figure(figsize=(12, 8))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 3. Explore Numerical Features
numerical_features = numeric_df.columns
for feature in numerical_features:
    if feature != 'Churn':  # Skip the target variable
        plt.figure(figsize=(8, 4))
        sns.boxplot(x='Churn', y=feature, data=df)
        plt.title(f'{feature} vs Churn')
        plt.show()

# 4. Explore Categorical Features
# Select non-numeric columns
categorical_features = df.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=feature, hue='Churn', data=df)
    plt.title(f'{feature} Distribution by Churn')
    plt.legend(title='Churn', loc='upper right')
    plt.show()

# 5. Additional Visualizations (if needed)
# Example: Tenure vs Churn
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, bins=30)
plt.title('Tenure Distribution by Churn')
plt.show()

# Example: MonthlyCharges vs Churn
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='MonthlyCharges', hue='Churn', kde=True, bins=30)
plt.title('Monthly Charges Distribution by Churn')
plt.show()
```
